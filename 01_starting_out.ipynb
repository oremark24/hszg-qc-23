## 1.1 A first quantum circuit

Discuss the following Qiskit code (next cell) and represented circuit.
1. What are the parts of the circuit (how does it compare to classical algorithms)?
2. What is the expected behavior?
3. Add quantum gates to execute simultaneous toss of 2 coins.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

qc = QuantumCircuit(2,2)

######################
# put your code here #
######################

qc.measure(0,0)
qc.measure(1,1)
display(qc.draw(output='mpl'))

backend = AerSimulator(shots=1024)
job = backend.run(qc)
result = job.result()
counts = result.get_counts()
display(plot_histogram(counts))
```

## 1.2 Write your initials

The following quantum circuit outputs the ascii codes of your instructor's initials.
1. Interpret the measured output.
2. Explain the behavior of the different gates and the circuit as a whole.
3. Modify the circuit to output your initials.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

qc = QuantumCircuit(8,8)
qc.x(0)
qc.h(1)
qc.x(2)
qc.cx(1,3)
qc.x(3)
qc.cx(1,4)
qc.x(6)
qc.barrier()
qc.measure(range(8),range(8))
display(qc.draw(output='mpl'))

backend = AerSimulator(shots=1024)
job = backend.run(qc)
result = job.result()
counts = result.get_counts()
display(plot_histogram(counts))
```
